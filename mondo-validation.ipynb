```python
dist = pd.read_csv("go_dist_only.csv", sep=',', names=['super','sub','distance'])
```

```python
dist['super'] = dist['super'].apply(lambda s: re.sub(r'^http://purl\.obolibrary\.org/obo/','',s))
dist['sub'] = dist['sub'].apply(lambda s: re.sub(r'^http://purl\.obolibrary\.org/obo/','',s))
```

```python
labels = list(sorted(set(list(dist['super'])+list(dist['sub']))))
label2idx = {}
for idx, val in enumerate(labels):
    label2idx[val]=idx
```

```python
X = scipy.sparse.dok_matrix((len(labels),len(labels)), dtype='int32')
for index, row in dist.iterrows():
    X[label2idx[row['super']],label2idx[row['sub']]] = row['distance']
    X[label2idx[row['sub']],label2idx[row['super']]] = row['distance']
```

```python
X
```

```python
pd.DataFrame(X,index=labels,columns=labels).to_csv('three_merged_dist_wide_bi.csv',sep=',')
```

```python
from sklearn.cluster import DBSCAN

# Compute DBSCAN
db = DBSCAN(eps=10, min_samples=10, metric='precomputed', n_jobs=72).fit(X)
```

```python
db.core_sample_indices_
```

```python
from collections import OrderedDict
pd.set_option('display.max_rows', 5000)
clusters = pd.DataFrame(OrderedDict(label=labels,cluster=db.labels_))
groups = OrderedDict()
for index, row in clusters.iterrows():
    label = re.sub(r'_.*$','',row['label'])
    if (label,row['cluster']) not in groups: groups[(label,row['cluster'])]=0
    groups[(label,row['cluster'])] += 1
groups2 = OrderedDict(label=[], cluster=[], members=[])
for k, v in groups.items():
    groups2['label'].append(k[0])
    groups2['cluster'].append(k[1])
    groups2['members'].append(v)
pd.DataFrame(groups2).sort_values(by=['label','cluster','members'], ascending=[True, True, False])
```

```python
d=OrderedDict(cluster=[], name=[], label=[])
for idx in db.core_sample_indices_:
    d['cluster'].append(db.labels_[idx])
    d['name'].append(labels[idx])
    d['label'].append(re.sub(r'_.*$','',labels[idx]))
pd.DataFrame(d).sort_values(by=['cluster','label','name']).to_csv('core_samples.csv',sep=',')
```
